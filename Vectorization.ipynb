```python
def gender_features(word):
    return {'last_letter': word[-1]}
```

```python
gender_features('Trump')
```

```python
from nltk.corpus import names
```

```python
names.words()
```

```python
len(names.words())
name_labels = ([(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')])
import random
random.shuffle(name_labels)
print(name_labels)
```

```python
featureSets = [(gender_features(n), gender) for (n,gender) in name_labels]
print(featureSets)
```

```python
train_set, test_set = featureSets[0:5000], featureSets[5001:]
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
classifier.classify(gender_features('david'))
```

```python
train_set, test_set = featureSets[0:5000], featureSets[5001:]
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
classifier.classify(gender_features('Natasha'))
```

```python
train_set, test_set = featureSets[0:5000], featureSets[5001:]
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
classifier.classify(gender_features('Gurpreet'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(binary=True)
corpus=["Tessaract is good optical character recognition engine  ", "optical character recognition is significant"]
```

```python
vect.fit(corpus)
```

```python
vocab=vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key, vocab[key]))
```

```python
print(vect.transform(["He am the smartest human being on this planet"]).toarray())
```

```python
print(vect.transform(["This is a good optical illusion"]).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect.transform(["Google Cloud Vision is a character recognition engine"]).toarray(), vect.transform(["OCR is an optical character recognition engine"]).toarray())
print(similarity)
```
